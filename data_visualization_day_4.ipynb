### Univariate analysis for Categorical dataset

## pie chart for categorical columns

#### column-cut

```python
x=df['cut'].value_counts().plot(kind = 'pie')
```

```python
labels = list(df['cut'].value_counts().index)
labels
```

```python
plt.pie(df['cut'].value_counts(),shadow=True,autopct='%0.2f%%',labels=labels)
plt.show()
```

```python
## here , we can see  diamonds ideal cut precntage found to be more then very good comes at second number,,
```

### column-clarity

```python
labels = list(df['clarity'].value_counts().index)
labels
```

```python
plt.pie(df['clarity'].value_counts(),shadow=True,autopct='%0.2f%%',labels=labels)
plt.show()
```

```python
## we got diamonds clarity of VS2 prcentage higher than others clarity,,
```

## categorical to categorical

### using crosstab

```python
df1 = pd.crosstab(df['cut'], df['clarity'])
df1
```

```python
df1.plot(kind = 'bar')
```

```python
## here we see that, more  diamonds clarity of VS2 type and even other clarity types are found to be of ideal cut diamonds ,then premium and very good cut has somehow same clarity of diamonds,,
```

```python
df1 = pd.crosstab(df['cut'], df['color'])
df1
```

```python
df1.plot(kind = 'bar')
```

```python
## here we see that, more  diamonds G color  and even other colors types are found to be of ideal cut diamonds ,then premium and very good cut has somehow same color of diamonds in number,
```
