######1. load the iris.csv dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
df=pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.info()
```

#### 1. create a bar plot for count the frequency of the three species.

```python
### code here
sns.countplot(df.Species)
plt.xticks(rotation = 90);
plt.title('Counting of Species in Iris_Dataset');
```

```python
##from this plot ,we can observe that Iris has three kind of species which are equal in number that is 50.
```

##### 2. Create a pie chart to show the percentage of species.

```python
df['Species'].value_counts().plot(kind = 'pie')
```

```python
lables = list(df['Species'].value_counts().index)
lables
```

```python
### code here
plt.pie(df['Species'].value_counts(),shadow=True,autopct='%0.2f%%',labels=lables)
plt.title('Prcentage of Species')
plt.show()
```

```python
### from here ,we get the equal prcentage of all three species.
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
### code here
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df, size=10,ratio=10, kind='scatter',color='green')
plt.show()
```

```python
## from this,we don't get any particular relation between the two variables used.
```

```python
sns.jointplot(x="PetalLengthCm", y="PetalWidthCm", data=df, size=10,ratio=10, kind='scatter',color='green')
plt.show()
```

##### 4. Create a scatter plot and take hue = "Species"

```python
### code here
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='Species',size=4).map(plt.scatter,'PetalLengthCm','PetalWidthCm').add_legend()
plt.show()
```

```python
## by this plot,we got that Iris-setosa species has less petal length and width as compare to the other species. and iris virginca has maximum petal length and width.
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='Species',size=4).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
plt.show()
```
