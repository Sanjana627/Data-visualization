##### 2. What is box plot ?

```python
### write here
It displays the five essential summary of set of data , these are minimum, first quartile, median, third quartile,and maximum .In box plot , we have a box which ranges from first quartile to third quartile and it contains median which is a vertical line in that particular box.And it has whishkers along its both sides ,which contains 25% of data respectively ,and box contains 50 % of data ,thus,whole box plot is divided into three groups which contains precentage of data.with the help of box plot , we can have maximum and minimum value of the dataset and we can calculate range of data, and the second advantage of box plot is to detect outliers.
```
